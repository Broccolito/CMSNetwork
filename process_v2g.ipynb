```python
import pandas as pd
import os
import glob
```

```python
def parquet_to_csv(parquet_path):
    d = pd.read_parquet(parquet_path)
    d["combined"] = d["chr_id"].astype(str) + "_" + d["position"].astype(str) + "_" + d["gene_id"].astype(str)
    d = d.drop_duplicates(subset="combined", keep="first")
    d = d[["chr_id", "position", "ref_allele", "alt_allele", "gene_id", "overall_score"]].copy()
    d.rename(columns={"chr_id": "chr"}, inplace=True)
    csv_path = parquet_path.replace(".parquet", ".csv")
    d.to_csv(csv_path, index=False)
    print(f"Saved processed file to: {csv_path}")
```

```python
for file_path in glob.glob("*.parquet"):
    parquet_to_csv(file_path)
```
